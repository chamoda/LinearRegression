```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rng = np.random.RandomState(42)
x = rng.rand(100) * 10
x = x[:, np.newaxis]
b = rng.rand(100) * 5
b = b[:, np.newaxis]
y = 3 * x + b
```

```python
plt.scatter(x, y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
def h(x, theta0, theta1):
    return theta0 + theta1 * x
```

```python
def cost(x, theta0, theta1):
    return np.sum(np.power(h(x, theta0, theta1) - y, 2)) * 1 / 2 * x.shape[0]
```

Trying to solve without gradient descent

```python
theta0 = np.linspace(0, 5, 500)
theta1 = np.linspace(0, 5, 500)
T0, T1 = np.meshgrid(theta0, theta1)
```

```python
theta0 = np.linspace(0, 5, 500)
theta1 = np.linspace(0, 5, 500)

T0, T1 = np.meshgrid(theta0, theta1)

C = np.zeros((500, 500))

for i in range(0, 500):
    for j in range(0, 500):
        C[i, j] = cost(x, T0[i, j], T1[i, j])
```

```python
plt.contourf(T0, T1, C)
plt.colorbar()
plt.show()
```
